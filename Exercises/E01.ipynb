```python
import pandas as pd

file = "student-mat-missing-data.csv"

df = pd.read_csv(file)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def barplot_with_missing_data(dataframe):
    df_to_plot = pd.DataFrame({"Missing values": dataframe.isnull().sum()})
    df_to_plot["Column name"] = dataframe.columns

    fig, ax = plt.subplots(1, 1, figsize = (12,4), dpi = 100)
    plt.xticks(rotation = 90)
    
    sns.barplot(data = df_to_plot, x = df_to_plot["Column name"], y = df_to_plot["Missing values"])
```

```python
from data_utils import barplot_with_missing_data

barplot_with_missing_data(df)
```

```python
import random as rnd
import matplotlib.pyplot as plt

rnd.seed(42)

days = list(range(50))
days_ripple = [rnd.uniform(0, 6) for _ in range(50)]
price = [700 + (1.7) ** (x / 3) + (y + x / 4) ** 2 for (x, y) in zip(days, days_ripple)]

tesla_df = pd.DataFrame({"Dag": days, "Pris": price})

fig, ax = plt.subplots(1, 1, figsize = (12,4), dpi = 100)
ax.set_ylim(0,7000)

sns.lineplot(data = tesla_df, x = "Dag", y = "Pris")
```
