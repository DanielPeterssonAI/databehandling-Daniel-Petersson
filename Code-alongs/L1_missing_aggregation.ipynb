## Missing data aggregations

```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 10

# create dummy data

random_matrix = np.random.randint(1, 10, [n, n])

index = np.random.choice(random_matrix.size, 10, replace = False)
print(index)

random_matrix = random_matrix * 1.0

random_matrix.ravel()[index] = None

print(random_matrix)
```

```python
scores = pd.DataFrame(random_matrix, 
                      index = [f"Player {i}" for i in range(1, n+1)],
                      columns = [f"Round {i}" for i in range(1, n+1)]
                      )

scores
```

```python
scores.info()
```

## Missing value operations
- isnull - returns True if null
- notnull - returns True if not null
- dropna() - drops an axis with nulls
- fillna() - fills the null values with certain values


```python
scores.isnull()
```

```python
scores.notnull()
```

```python
scores.dropna()
```

```python
scores.dropna(axis = 1)
#axis = 1 är samma som axis = "columns"
```

```python
# How to treat NaNs?
# depends on the situations
# smany times need to talk to domain expert

scores.fillna(0) # in this case - may be reasonable to give 0 scores for missing values
```

## Missing value strategy

Strategy depends on:
- dataset size
- valuable information
- percentage missing values
- domain knowledge
- missing values can impact
    - data visualization
    - artihmetic computation
    - summary statistics
    - machine learning algorithms

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
# focus on age

sns.histplot(data = titanic, x = "age", kde = True, hue = "sex", alpha = 1)
```

```python
days = [i for i in range(0, 20)]
price = [i ** (0.1 * i) for i in days]

tesla_df = pd.DataFrame({"Dag": days, "Pris": price})


#sns.lineplot(data = tesla_df, x = "Dag", y = "Pris")
sns.barplot(data = tesla_df, x = "Dag", y = "Pris")
```

```python
# talked to a titanic historian
# this dude says: use median age of corresponding gender to fill ages

print(f"Number of males: {np.sum(titanic['sex'] == 'male')}")
print(f"Number of females: {np.sum(titanic['sex'] == 'female')}")

#titanic.loc[titanic["sex"] == "male", "age"]

median_male_age = titanic.loc[titanic["sex"] == "male", "age"].median()
median_female_age = titanic.loc[titanic["sex"] == "female", "age"].median()

median_female_age, median_male_age
```

```python
titanic.loc[titanic["sex"] == "male", "age"] = titanic.loc[titanic["sex"] == "male", "age"].fillna(median_male_age)
```

```python
titanic
```

```python
titanic.loc[titanic["sex"] == "female", "age"] = titanic.loc[titanic["sex"] == "female", "age"].fillna(median_female_age)
```
